```python
#Speed up of cpu increase
#'/content/Light_nproc.csv', '/content/Heavy_nproc.csv', '/content/Heavy_ncpu_inc.csv', 
# List of file names
files = ['/content/Heavy_cpu_incmean.csv']

# Read each file and store the dataframe in the list

HNCPU=pd.read_csv('/content/Heavy_cpu_incmean.csv')
# Calculate speed up
HNCPU['speed_up'] = HNCPU['total_time'].iloc[0] / HNCPU['total_time']

plt.title(' NCPU change  v speedup')
plt.ylabel('speed up')
plt.xlabel('Number of processes')
plt.plot(HNCPU['size'], HNCPU['speed_up'])
plt.show()

```
